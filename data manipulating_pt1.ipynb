```python
import pandas as pd
dcr = pd.read_csv("/content/drive/MyDrive/Data/Clash_stats.zip")
dcr.head(3)
```

```python
#drop buildings and spells
dcr = dcr[dcr['Type']=='Troops and Defenses']
dcr = dcr.drop([6, 7, 10, 17, 20, 23, 27, 29, 30, 33, 40, 51, 56]) 
```

```python
x = list(dcr)
nas = []

for atts in x:
  nas.append(atts)
  nas.append(dcr[atts].isna().sum())

nas
```

```python
#drop irrelevant columns
dcr = dcr.drop(['Spawn Damage', 
                'Crown Tower Damage',
                'Card Level (Spawn Level)',
                 'Maximum Spawned',
                 'Radius',  
                 'Spawn DPS',
                 'Spawn Damage',
                 'Spawn Health',
                 'Spawn Speed',
                 'Spawner Health',
                 'Troop Spawned', 
                 'Type'], 
                 axis=1)
```

```python
dcr.isna().sum().sum()
```

```python
#make corrections
dcr.at[11, 'Health (+Shield)'] = 1392
dcr.at[12, 'Health (+Shield)'] = 1243
dcr.at[24, 'Health (+Shield)'] = 285
```

```python
#define speeds
fast = ['Mini P.E.K.K.A.', 'Bandit', 'Electro Wizard', 'Miner']
slow = ['Royal Giant', 'Giant', 'P.E.K.K.A.', 'Golem', 'Lava Hound', 'Bowler', 'Ice Golem', 'Sparky']
very_fast = ['Bats', 'Spear Goblins', 'Goblins', 'Elite Barbarians', 'Hog Rider', 'Dart Goblin', 'Lumberjack']

dcr.insert(10, "Speed", 'Medium')

for cards in fast:
  dcr.loc[(dcr.Card == cards),'Speed']='Fast'
for cards in slow:
  dcr.loc[(dcr.Card == cards),'Speed']='Slow'
for cards in very_fast:
  dcr.loc[(dcr.Card == cards),'Speed']='Very Fast'
```

```python
#define type 
magic_troop = ['Electro Wizard', 'Ice Wizard', 'Night Witch', 'Witch', 'Wizard']
creature = ['Bowler', 'Dart Goblin', 'Goblins', 'Spear Goblins']
magic_creature = ['Golem', 'Ice Golem']
flying_creature = ['Baby Dragon', 'Lava Hound', 'Minions', 'Minion Horde', 'Mega Minion', 'Bats']
vehicle = ['Sparky', 'Cannon Cart', 'Balloon']

dcr.insert(10, "Type", 'Troop')

for cards in magic_troop:
  dcr.loc[(dcr.Card == cards),'Type']='Magic Troop'
for cards in creature:
  dcr.loc[(dcr.Card == cards),'Type']='Creature'
for cards in magic_creature:
  dcr.loc[(dcr.Card == cards),'Type']='Magic Creature'
for cards in flying_creature:
  dcr.loc[(dcr.Card == cards),'Type']='Flying Creature' 
for cards in vehicle:
  dcr.loc[(dcr.Card == cards),'Type']='Vehicle'
```

```python
#define target
building = ['Hog Rider', 'Royal Giant', 'Golem', 'Giant', 'Lava Hound', 'Ice Golem', 'Balloon']
groundair = ['Eletcro Wizard', 'Baby Dragon', 'Minions', 'Minion Horde', 'Mega Minion', 'Executioner', 'Musketeer', 'Wizard', 'Witch', 'Ice Wizard', 'Princess', 'Archers', 'Three Musketeers', 'Dart Goblin', 'Spear Goblin', 'Bats']

dcr.insert(10, "Targets", 'Ground')

for cards in building:
  dcr.loc[(dcr.Card == cards),'Targets']='Buildings' 
for cards in groundair:
  dcr.loc[(dcr.Card == cards),'Targets']='Ground & Air'
```

```python
dcr
```

```python
#define deploy time
three = ['P.E.K.K.A.', 'Golem']

dcr.insert(10, "Deploy Time", 1)

dcr.loc[(dcr.Card == 'Royal Giant'),'Deploy Time']=2
for cards in three:
  dcr.loc[(dcr.Card == cards),'Deploy Time']= 3
```

```python
#rename columns
dcr.rename(columns={'Health (+Shield)': 'Health'}, inplace=True)
```

```python
#rearrange columns
dcr = dcr[['Card', 'Cost','Count','Health','Damage','Speed','Range', 'Deploy Time', 'Hit Speed','Damage per second','Death Damage','Targets', 'Type']]
```

```python
#sort dtypes regarding use of comma
dcr['Health'] = dcr['Health'].astype(str)
dcr['Damage'] = dcr['Damage'].astype(str)

dcr['Health'] = dcr['Health'].str.replace(',','')
dcr['Damage'] = dcr['Damage'].str.replace(',','')
```

```python
#sort dtypes
floats = ['Range', 'Hit Speed']
ints = ['Cost', 'Damage', 'Health', 'Count', 'Damage per second', 'Death Damage']

for atts in ints:
  dcr[atts] = dcr[atts].astype('int64')
for atts in floats:
  dcr[atts] = dcr[atts].astype('float64')
  
dcr.dtypes
```
